# Generar una señal senoidal y mostrar sus gráficas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# valores de la señal
A = 1
f = 100           #  hz
w = 2*np.pi*f     #  rad/s
phi = 0

fs = 10*f
Ts = 1/fs

n = np.arange(1000)
t = n*Ts

# señal senoidas
x = A*np.sin(w*t + phi)
```

```python
#plt.plot(t, x)
plt.stem(n, x)
#plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal Senoidal')
plt.xlim(0,100)
```

## FFT

```python
plt.magnitude_spectrum(x, Fs=fs, scale='dB')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud (dB)')
plt.grid(ls=":")
```

## STFT

```python
from scipy import signal
```

```python
f, t, Zxx = signal.stft(x, fs, nperseg=100)

# Graficar el espectrograma
plt.figure(figsize=(15, 4))
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('Espectrograma de la STFT')
plt.ylabel('Frecuencia [Hz]')
plt.xlabel('Tiempo [s]')
plt.colorbar(label='Magnitud')
plt.show()
```

## Ejercicio
Suma 3 señales diferentes, muestreala y muestra sus gráfica, FFT y SFTT

```python
#Parametros de muestreo
fs = 1000
Ts = 1/fs
t = np.arange(0,1,Ts)

# Señales
s1 = np.sin(2*np.pi*10*t)      # 10 Hz
s2 = 0.5*np.sin(2*np.pi*50*t)  # 50 Hz
s3 = np.sin(2*np.pi*100*t)  # 100 Hz

s = s1 + s2 + s3
```

```python
#Gráfica
plt.plot()
plt.plot(t, s)
plt.title("Señal sumada")

plt.xlabel("Tiempo (s)")
plt.xlim(0,1)
plt.grid(ls=":")
```

### FFT

```python
plt.magnitude_spectrum(s, Fs=fs, scale='dB')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud (dB)')
plt.xlim(0,300)
plt.grid(ls=":")
```

### STFT

```python
f, t, Zxx = signal.stft(s, fs, nperseg=256)

# Graficar el espectrograma
plt.figure(figsize=(15, 4))
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('Espectrograma de la STFT')
plt.ylabel('Frecuencia [Hz]')
plt.xlabel('Tiempo [s]')
plt.colorbar(label='Magnitud')
plt.ylim(0,200)
plt.show()
```
